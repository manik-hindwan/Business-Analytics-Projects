### 3. Cohort analysis

```python
# Creating a dataframe with only relevant columns since InvoiceNo has duplicate entries for a customer due to many items 

print(df.columns)
df_cohort = df[['CustomerID','InvoiceNo','InvoiceDate']].drop_duplicates().reset_index(drop=True)
```

```python
df_cohort.head(2)
```

```python
# Creating a custom column for the period month

df_cohort['order_month'] = df_cohort.InvoiceDate.dt.to_period('M')
df_cohort.head(2)
```

```python
# Creating a column which has a cohort value for every customer

df_cohort['cohort'] = df_cohort.groupby('CustomerID')['InvoiceDate'].transform('min').dt.to_period('M')
df_cohort.head(2)
```

```python
# Counting the number of unique customers belonging to a particular cohort

df_cohort = df_cohort.groupby(['cohort','order_month']).agg(n_customers=('CustomerID','nunique')).reset_index(drop=False)
df_cohort.head(3)
```

```python
# Assigning each cohort a number on the basis of cohort period and order_month

df_cohort['cohort_number'] = (df_cohort.order_month - df_cohort.cohort).apply(attrgetter('n'))
df_cohort.head(4)
```

```python
# Create a customer cohort pivot

cohort_pivot = df_cohort.pivot_table(values = 'n_customers', 
                                     index = 'cohort', 
                                     columns = 'cohort_number')
cohort_pivot
```

```python
cohort_size = cohort_pivot.iloc[:,0]
retention_matrix = cohort_pivot.divide(cohort_size,axis=0)
retention_matrix
```

### 4. Plotting the cohort

```python
with sns.axes_style("white"):
    fig, ax = plt.subplots(1, 2, figsize=(12, 8), sharey=True, gridspec_kw={'width_ratios': [1, 11]})
    
    # retention matrix
    sns.heatmap(retention_matrix, 
                mask=retention_matrix.isnull(), 
                annot=True, 
                fmt='.0%', 
                cmap='RdYlGn', 
                ax=ax[1])
    ax[1].set_title('Monthly Cohorts: User Retention', fontsize=16)
    ax[1].set(xlabel='# of periods',
              ylabel='')

    # cohort size
    cohort_size_df = pd.DataFrame(cohort_size).rename(columns={0: 'cohort_size'})
    white_cmap = mcolors.ListedColormap(['white'])
    sns.heatmap(cohort_size_df, 
                annot=True, 
                cbar=False, 
                fmt='g', 
                cmap=white_cmap, 
                ax=ax[0])

    fig.tight_layout()
```
